```python
df_18.groupby('fuel').mean()
```

```python
df_18.groupby('veh_class').mean()
```

Vehicle classes improved in fuel economy (increased in mpg) as under:

```python
df_08.groupby('veh_class').mean()['cmb_mpg'].hist()
plt.xlabel('Vehicle class')
plt.ylabel('Combined mpg')
plt.title('Vehicle Class vs Combined MPG');
```

### Q3: What are the characteristics of SmartWay vehicles? Have they changed over time?

```python
df_08.groupby('smartway').mean()
```

```python
df_18.groupby('smartway').mean()
```

```python
df_08.groupby('smartway').mean()['air_pollution_score'].hist()
```

```python
df_18.groupby('smartway').mean()['air_pollution_score'].hist()
```

```python
df_08.groupby('smartway').mean()['greenhouse_gas_score'].hist()
```

```python
df_18.groupby('smartway').mean()['greenhouse_gas_score'].hist()
```

```python
ggs_08= df_08.groupby('smartway').mean()['greenhouse_gas_score']
ggs_08_label= ['No', 'Yes']

plt.pie(ggs_08, labels=ggs_08_label);
```

```python
ggs_18= df_18.groupby('smartway').mean()['greenhouse_gas_score']
ggs_18_label= ['Elite','No', 'Yes']

plt.pie(ggs_18, labels=ggs_18_label);
```

```python
aps_08= df_08.groupby('smartway').mean()['air_pollution_score']
aps_08_label= ['No', 'Yes']

plt.pie(aps_08, labels=aps_08_label, colors='gk');
```

```python
aps_18= df_18.groupby('smartway').mean()['air_pollution_score']
aps_18_label= ['Elite','No', 'Yes']

plt.pie(aps_18, labels=aps_18_label,colors='rgk');
```

**Q3: What are the characteristics of SmartWay vehicles? Have they changed over time? (mpg, greenhouse gas)**
- Characteristics are Air pollution score and Greenhouse Gas Score
- These have improved from 2008 to 2018.
- Air pollution scores reduced and Greenhouse scores increased.


### Q4: What features are associated with better fuel economy?

```python
df_08.corr('pearson')
```

According to the 2008 dataset, we see that better fuel economy is associated with the following factors as [using correlation]:
- Negatively with displ : -0.818799
- Negatively with cyl : -0.770538
- Positively with Greenhouse gas scores : 0.947607

```python
df_18.corr('pearson')
```

According to the 2018 dataset, we see that better fuel economy is associated with the following factors as [using correlation]:
- Positively with Greenhouse gas scores : 0.814982

Based on both the datasets, we can associate Greenhouse Gas Scores as being positively associated with Fuel Economy.
